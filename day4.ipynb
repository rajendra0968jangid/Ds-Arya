**Exception Handling**
1. try, except, finally

```python
# try except
try:
  num1  = 10
  num2 = 5
  print(num1/num2)
except:
  print("hello except")

# try except
try:
  num1 = 10
  num2 = 0
  print(num1/num2)
except:
  print("hello except")

# try except and final
try:
  num1 = 10
  num2 = 5
  print(num1/num2)
except:
  print("hello except")
finally:
  print("hello finally divide")
```
