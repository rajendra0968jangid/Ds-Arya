**Operation on Rows**

2. print(df.loc[row]) -> label based (get complete row)
3. print(df.loc[[0, 2]]) -> get multiple rows
4. print(df.loc[[0, 2], ["col1", "col2"]]) multi rows and cols

**Filter DataFrame**
1. Filter DataFrame -> Ex. df[df["Age"] > 22]
2. Filter Using Greater Than (>)
3. df[df["column"] > value]
4. Filter Using Less Than (<)
5. df[df["column"] < 22]
6. Filter Using Equal To (==)
7. df[df["column"] == 22]
8. Filter Using Not Equal To (!=)
9. df[df["Age"] != 22]
10. Filter String Columns
11. df[df["column"] == "Sara"]
12. Multiple Conditions (AND)
13. df[(df["col1"] > 22) & (df["col2"] > 55000)]
14. Multiple Conditions (OR)
15. df[(df["col1"] > 23) | (df["col2"] > 55000)]
16. Using query()
17. df.query("col1 > 22")
